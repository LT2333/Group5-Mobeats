```python
import os
import numpy as np
```

```python
dir_path = './Train Data/' #总的文件夹名称
out_path = './Train Data Augmentation' #输出的文件夹名称 #Data Augmentation
```

```python
# 创建输出文件夹
if not os.path.exists(out_path):
    os.makedirs(out_path)
# 六种输出文件夹路径
out_path1 = out_path + '/Zoom In'
out_path2 = out_path + '/Zoom Out'
out_path3 = out_path + '/Left'
out_path4 = out_path + '/Right'
out_path5 = out_path + '/CW Rotate 15'
out_path6 = out_path + '/CCW Rotate 15'
# 输出路径列表
out_list = []
out_list.append(out_path1)
out_list.append(out_path2)
out_list.append(out_path3)
out_list.append(out_path4)
out_list.append(out_path5)
out_list.append(out_path6)
# 创建六种输出文件夹
for i in out_list:
    if not os.path.exists(i):
        os.makedirs(i)
```

```python
# Zoom in
def zoom_in(x,y,ratio):
    return 0.5+(x-0.5)*ratio,0.5+(y-0.5)*ratio
```

```python
# Zoom out
def zoom_out(x,y,ratio):
    return 0.5+(x-0.5)/ratio,0.5+(y-0.5)/ratio
```

```python
# Left Shift
def left(x,y,ratio):
    return x-ratio,y
```

```python
# Right Shift
def right(x,y,ratio):
    return x+ratio,y
```

```python
import math
# CCW
def rl(x,y,ratio):
    ratio = ratio/180*math.pi
    r = ((x-0.5)**2 + (y-0.5)**2)**0.5
    xita = math.atan((y-0.5)/(x-0.5))
    alfa = xita + ratio
    return 0.5+r*math.cos(alfa),0.5+r*math.sin(alfa)
```

```python
import math
# CW
def rr(x,y,ratio):
    ratio = ratio/180*math.pi
    r = ((x-0.5)**2 + (y-0.5)**2)**0.5
    xita = math.atan((y-0.5)/(x-0.5))
    alfa = xita - ratio
    return 0.5+r*math.cos(alfa),0.5+r*math.sin(alfa)
```

```python
for i in os.listdir(dir_path):
    print(i)
    # 创建子文件夹
    for k in out_list:
        if not os.path.exists(k + '/' + i):
            os.makedirs(k + '/' + i)
    # 遍历每个npy
    for j in os.listdir(dir_path + '/' + i):
        print(dir_path + '/' + i + '/' + j)
        # 创建子子文件夹
        for k in out_list:
            if not os.path.exists(k + '/' + i + '/' + j):
                os.makedirs(k + '/' + i + '/' + j)
        # 遍历子子文件夹
        for ny in os.listdir(dir_path + '/' + i + '/' + j):
            if os.path.splitext(ny)[1] == '.npy': # 筛选npy文件
                arr=np.load(dir_path + '/' + i + '/' + j + '/' + ny)
                # 创建六个输出
                out1 = arr.copy()
                out2 = arr.copy()
                out3 = arr.copy()
                out4 = arr.copy()
                out5 = arr.copy()
                out6 = arr.copy()
                # 处理身体坐标
                for n in range(33):
                    x = arr[0 + n*4]
                    y = arr[1 + n*4]
                    # 放大
                    x1,y1 = zoom_in(x,y,ratio=1.2) # ratio 为放大比例
                    # 缩小
                    x2,y2 = zoom_out(x,y,ratio=1.2) # ratio 为缩小比例
                    # 左移
                    x3,y3 = left(x,y,ratio=0.1) # ratio 为左移的距离
                    # 右移
                    x4,y4 = right(x,y,ratio=0.1) # ratio 为右移的距离
                    # 顺时针旋转
                    x5,y5 = rr(x,y,ratio=15) # ratio 为旋转的角度角°
                    # 逆时针旋转
                    x6,y6 = rl(x,y,ratio=15) # ratio 为旋转的角度角°
                    # 保存结果
                    out1[0 + n*4] = min(max(x1,0),1)
                    out1[1 + n*4] = min(max(y1,0),1)
                    out2[0 + n*4] = min(max(x2,0),1)
                    out2[1 + n*4] = min(max(y2,0),1)
                    out3[0 + n*4] = min(max(x3,0),1)
                    out3[1 + n*4] = min(max(y3,0),1)
                    out4[0 + n*4] = min(max(x4,0),1)
                    out4[1 + n*4] = min(max(y4,0),1)
                    out5[0 + n*4] = min(max(x5,0),1)
                    out5[1 + n*4] = min(max(y5,0),1)
                    out6[0 + n*4] = min(max(x6,0),1)
                    out6[1 + n*4] = min(max(y6,0),1)
                # 处理脸的坐标
                for n in range(468):
                    x = arr[132 + n*3]
                    y = arr[133 + n*3]
                    # 放大
                    x1,y1 = zoom_in(x,y,ratio=1.2) # ratio 为放大比例
                    # 缩小
                    x2,y2 = zoom_out(x,y,ratio=1.2) # ratio 为缩小比例
                    # 左移
                    x3,y3 = left(x,y,ratio=0.1) # ratio 为左移的距离
                    # 右移
                    x4,y4 = right(x,y,ratio=0.1) # ratio 为右移的距离
                    # 顺时针旋转
                    x5,y5 = rr(x,y,ratio=15) # ratio 为旋转的角度角°
                    # 逆时针旋转
                    x6,y6 = rl(x,y,ratio=15) # ratio 为旋转的角度角°
                    # 保存结果
                    out1[132 + n*3] = min(max(x1,0),1)
                    out1[133 + n*3] = min(max(y1,0),1)
                    out2[132 + n*3] = min(max(x2,0),1)
                    out2[133 + n*3] = min(max(y2,0),1)
                    out3[132 + n*3] = min(max(x3,0),1)
                    out3[133 + n*3] = min(max(y3,0),1)
                    out4[132 + n*3] = min(max(x4,0),1)
                    out4[133 + n*3] = min(max(y4,0),1)
                    out5[132 + n*3] = min(max(x5,0),1)
                    out5[133 + n*3] = min(max(y5,0),1)
                    out6[132 + n*3] = min(max(x6,0),1)
                    out6[133 + n*3] = min(max(y6,0),1)
                # 处理左手的坐标
                for n in range(21):
                    x = arr[1536 + n*3]
                    y = arr[1537 + n*3]
                    # 放大
                    x1,y1 = zoom_in(x,y,ratio=1.2) # ratio 为放大比例
                    # 缩小
                    x2,y2 = zoom_out(x,y,ratio=1.2) # ratio 为缩小比例
                    # 左移
                    x3,y3 = left(x,y,ratio=0.1) # ratio 为左移的距离
                    # 右移
                    x4,y4 = right(x,y,ratio=0.1) # ratio 为右移的距离
                    # 顺时针旋转
                    x5,y5 = rr(x,y,ratio=15) # ratio 为旋转的角度角°
                    # 逆时针旋转
                    x6,y6 = rl(x,y,ratio=15) # ratio 为旋转的角度角°
                    # 保存结果
                    out1[1536 + n*3] = min(max(x1,0),1)
                    out1[1537 + n*3] = min(max(y1,0),1)
                    out2[1536 + n*3] = min(max(x2,0),1)
                    out2[1537 + n*3] = min(max(y2,0),1)
                    out3[1536 + n*3] = min(max(x3,0),1)
                    out3[1537 + n*3] = min(max(y3,0),1)
                    out4[1536 + n*3] = min(max(x4,0),1)
                    out4[1537 + n*3] = min(max(y4,0),1)
                    out5[1536 + n*3] = min(max(x5,0),1)
                    out5[1537 + n*3] = min(max(y5,0),1)
                    out6[1536 + n*3] = min(max(x6,0),1)
                    out6[1537 + n*3] = min(max(y6,0),1)
                # 处理右手的坐标
                for n in range(21):
                    x = arr[1599 + n*3]
                    y = arr[1600 + n*3]
                    # 放大
                    x1,y1 = zoom_in(x,y,ratio=1.2) # ratio 为放大比例
                    # 缩小
                    x2,y2 = zoom_out(x,y,ratio=1.2) # ratio 为缩小比例
                    # 左移
                    x3,y3 = left(x,y,ratio=0.1) # ratio 为左移的距离
                    # 右移
                    x4,y4 = right(x,y,ratio=0.1) # ratio 为右移的距离
                    # 顺时针旋转
                    x5,y5 = rr(x,y,ratio=15) # ratio 为旋转的角度角°
                    # 逆时针旋转
                    x6,y6 = rl(x,y,ratio=15) # ratio 为旋转的角度角°
                    # 保存结果
                    out1[1599 + n*3] = min(max(x1,0),1)
                    out1[1600 + n*3] = min(max(y1,0),1)
                    out2[1599 + n*3] = min(max(x2,0),1)
                    out2[1600 + n*3] = min(max(y2,0),1)
                    out3[1599 + n*3] = min(max(x3,0),1)
                    out3[1600 + n*3] = min(max(y3,0),1)
                    out4[1599 + n*3] = min(max(x4,0),1)
                    out4[1600 + n*3] = min(max(y4,0),1)
                    out5[1599 + n*3] = min(max(x5,0),1)
                    out5[1600 + n*3] = min(max(y5,0),1)
                    out6[1599 + n*3] = min(max(x6,0),1)
                    out6[1600 + n*3] = min(max(y6,0),1)
                # 保存npy
                np.save(out_path1 + '/' + i + '/' + j + '/' + ny,out1)
                np.save(out_path2 + '/' + i + '/' + j + '/' + ny,out2)
                np.save(out_path3 + '/' + i + '/' + j + '/' + ny,out3)
                np.save(out_path4 + '/' + i + '/' + j + '/' + ny,out4)
                np.save(out_path5 + '/' + i + '/' + j + '/' + ny,out5)
                np.save(out_path6 + '/' + i + '/' + j + '/' + ny,out6)
                
        
```
